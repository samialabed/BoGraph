```python
import matplotlib.pyplot as plt
import seaborn as sns
from autorocks.viz import plots_setup
import torch
import gpytorch
import botorch.posteriors
from botorch import models
from botorch.optim import fit
import torch
import numpy as np
import pandas as pd
import botorch
import daft

output_location = "/Users/salabed/workspace/latex_writings/thesis/phd_dissertation/Chapters/BoBn/Figures/"
plt.style.use("ggplot")
sns.set_theme(style="ticks", rc={"axes.spines.right": False, "axes.spines.top": False})
sns.set_context("paper")  # , font_scale=1.5, rc={"lines.linewidth": 1.5})
plt.rcParams["svg.fonttype"] = "none"
plt.rcParams["font.family"] = "Arial"
plt.rc("text", usetex=False)
plt.rc("xtick", labelsize="large")
plt.rc("ytick", labelsize="large")
plt.rc("axes", labelsize="large")
plt.rc("pdf", use14corefonts=True)

```

```python
# Instantiate the PGM.
pgm = daft.PGM()

pgm.add_node("x1", "$x_1$", 0, 0, scale=1)
pgm.add_node("x2", "$x_2$", 1, 0, scale=1)
pgm.add_node("model", "$f(x_1, x_2)$", 0.5, -1, scale=1.35)
pgm.add_node("branin", "Branin", -0.5, -1, scale = 1.25, observed = True)
pgm.add_node("currin", "Currin", 1.5, -1, scale = 1.25, observed = True)
pgm.add_edge(f"x1", "model")
pgm.add_edge(f"x2", "model")
pgm.add_edge(f"model", "currin")
pgm.add_edge(f"model", "branin")


# Render and save.
output_format = "pdf"
pgm.render(dpi=300).figure.savefig(f"{output_location}/mt_example.{output_format}", bbox_inches="tight", format=f"{output_format}", dpi=300)
```
